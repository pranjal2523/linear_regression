GRADIENT DESENT Implementation  


```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
X,Y
```

```python
# Goal - Given a Function f(x), we want to find the value of x that minimizes f(x)
#visualize function
plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('F(X)')
plt.show()
```

```python
plt.style.use('seaborn')
x = 0
lr = 0.1
error = []
plt.plot(X, Y)
for i in range(50):
    grad = 2*(x-5)
    x = x-lr*grad
    fx = (x-5)**2
    error.append(fx)
    plt.scatter(x, fx)
     
    
```

```python
#plot the value of errors 
plt.plot(error)
```
