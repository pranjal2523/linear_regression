```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Surface Plot

#a = np.array([1,2,3])
#b = np.array([4,5,6,7])
a = np.arange(-2,2,0.03)
b = a
a,b = np.meshgrid(a, b)

a, b
```

```python
plt.style.use("dark_background")
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a, b, a**2+b**2+10 , cmap="coolwarm")
plt.show()
```

```python
plt.style.use("classic")
fig = plt.figure() 
axes = fig.gca(projection='3d')
axes.contour(a, b, a**2+b**2+10 , cmap="rainbow")

plt.show()
```
